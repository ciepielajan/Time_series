Utworzenie DF  dla tygodniowych okresów (7 dniowych) w którym pierwsze 6 dni będzie X a  ostatni dzień (siódmy) y. 

```python
import numpy as np
import pandas as pd
```

wygenerowanie losowych danych (cena per godzina ) dla całego roku 
* żeby ułatwić weryfikacje poprawności transformacji wartość pm10 będą zwiększały się co 1

```python
time = pd.date_range(start="2021-01-1 00:00:00", end="2021-12-31 23:00:00", freq="1H")
data = range(1,len(time)+1) #np.random.randint(5,30,size=(len(time),1))

df = pd.DataFrame(data, index=time, columns=["pm10"])
print(df.shape)
df[:3]
```

```python
zakres = 168  # ostatnie x godzin 

for i in range(1, zakres + 1):
  i = i * -1

  df[f"pm10{i}"] = df["pm10"].shift(i)

df.dropna(inplace=True )  # usuwam nowo powstałe. 
df[:10]
```

```python
df[:3]
```

ustalam ile % całego zbioru ma stanowić zbiór testowy

```python
test_split = 0.2
```

```python
df_size = len(df)
train_size = df_size - int(test_split * df_size)
test_size = df_size - train_size

df_train = df.iloc[:train_size]   # granica zbioru train
df_test = df.iloc[train_size:df_size+1]   # granica zbioru test
```

dziele zbior na X_train, X_test, y_train, y_test


```python
X_train, X_test, y_train, y_test = df_train.iloc[:, 0:-2], df_test.iloc[:, 0:-2], df_train.iloc[:, -1], df_test.iloc[:, -1]
```
